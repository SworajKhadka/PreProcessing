```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
# NOTE: You MUST change this path to your actual file location!
df = pd.read_csv('Iris.csv')
```

```python
df
```

```python
X = df[["SepalLengthCm", "SepalWidthCm"]].values
y = df["Species"].replace({"setosa": 0, "versicolor": 1, "virginica": 2}).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(X_train_std, y_train)
```

```python
y_pred = svm.predict(X_test_std)
```

```python
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
data = {
    'document': ['I love this movie', 'This was an amazing play', 'I feel bored with this song', 'This show is a waste of time'],
    'class': ['positive', 'positive', 'negative', 'negative']
}
df = pd.DataFrame(data)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['document'])
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Initialize and train the Naïve Bayes Classifier
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```
