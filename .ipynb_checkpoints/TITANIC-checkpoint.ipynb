```python
print(f'the size of the dataset is :{df.shape}')
```

```python
df.drop(['body'], axis= 1, inplace= True)
print(f'the size of the dataset after dropping a body is {df.shape}')
```

```python
#We cant just use drop features as it will create a significant error in the production/determining the output hence we have to make sure the values are modified acording to the simple imputator freature we have in scikit learn
#it uses mean ,median and mode
from sklearn.impute import SimpleImputer
```

```python
print(f'the Number of null values before Imputing is :{df.age.isnull().sum()}')
```

```python
#The simple imputor has 3 main instances that it includes and they are: 1 Missing values, those are NAN values in our datasets. 2 strategy, the data which replaces the missing values aka strategical values via mean,median and mode, 3 fill values are the constant values to be filled
```

```python
imp=SimpleImputer(strategy='mean')
df['age']=imp.fit_transform(df[['age']])
print(f'the Number of null values after Imputing is:{df.age.isnull().sum()}')
```

```python
def get_parameters(df):
    parameters ={}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype =='float64' or df[col].dtype =='int64' or df[col].dtype=='int32':
            strategy='mean'
        else:
            strategy='most_frequent'
        missing_values = df[col][df[col].isnull()].values[0]
        parameters[col]= {'missing_values':missing_values,'strategy':strategy}
    return parameters
get_parameters(df)
```

```python
parameters =get_parameters(df)
```

```python
parameters
```

```python
parameters.items()
```

```python
for col, param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']
    imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```
